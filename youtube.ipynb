# PROJECT TITLE : top youtube channel

This dataset showcases the top 100 YouTube channels in 2024, including key metrics such as subscribers, total views, and video count. It has been manually curated to provide a snapshot of the leading channels as of December 2024. The data reflects the performance of channels across various genres, from entertainment to educational content, offering insights into the most influential creators of the year. This data is useful for marketers, content creators, and data scientists who want to analyze trends, audience engagement, and growth strategies of top YouTube influencers. By examining these metrics, users can gain valuable insights into content creation, viewer preferences, and platform dynamics. The dataset is based on publicly available YouTube statistics and third-party platforms, ensuring transparency and accuracy.

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# load the Dataset
data = pd.read_csv('YOUTUBE CHANNELS DATASET.csv')
```

```python
data
```

# data cleaning

```python
#Check for missing values 
print ("Missing Values:\n", data.isnull().sum())
```

```python
# Fill missing numerical values with the median 
data.fillna(data.median(numeric_only=True), inplace=True)
data
```

```python
# Drop duplicate rows
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
# Standardize column names and data formats(e.g., dates,numerical values)
data.columns = data.columns.str.strip().str.replace(' ','_')
```

```python
data
```

# Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of analyzing a dataset to summarize its main characteristics, uncover patterns, detect anomalies, test hypotheses, and check assumptions using statistical and graphical techniques. It is often the first step in any data analysis or data science project and helps you understand the data better before applying models or makingÂ decisions.

```python
# Summary statistics
data.describe()
```

```python
# Top 10 channels by Subscribers
top_channels = data.sort_values('Subscribers', ascending=True).head(10)

```

```python
print("Top 10 Channels by Subscribers:\n", top_channels[['Username', 'Subscribers']])
```

```python
# Bar chart for top channels
plt.figure(figsize=(10,6))
plt.barh(top_channels['Username'],top_channels['Subscribers'],color='skyblue')
plt.xlabel('Subscribers')
plt.title('Top 10 Channels by Subscribers')
plt.ylabel('Username')
plt.show()
```

```python
# correlation heatmap
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```

```python
# distribution of views
plt.figure(figsize=(8, 5))
sns.histplot(data['Views'], bins=30, kde=True, color='green')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```
